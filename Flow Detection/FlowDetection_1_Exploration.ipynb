```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/Users/yousefnami/KinKeepers/ProjectAI/Data/FlowSimulator/'
frames = []
for filename in os.listdir(path):
    if '.csv' in filename:
        frames.append(pd.read_csv(path+filename))

df = pd.concat(frames,ignore_index = True)

df.head()
```

```python
plt.figure(figsize = (32,32))
plt.plot(df.index,df.flow,'--')
```

```python
month = df[['2021-09-21' in i for i in df.date_time.values]]
plt.figure(figsize = (16,4))
month.reset_index(inplace = True)


plt.plot(month.index,month.flow,'--')
```
